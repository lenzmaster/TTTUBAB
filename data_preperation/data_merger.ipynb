```python
# Imports
import pandas as pd
import os.path
import numpy as np
import os
import glob
import sys

```

```python
# Load files - Assumption : files are clean

# Change to cleaned directory
#os.chdir('cleaned/')
filesGames = glob.glob('*games.pkl')
filesGameData = glob.glob('*game-data.pkl')
filesGames.sort()
filesGameData.sort()

gamesList = []
gameDataList = []

for i in range(len(filesGames)):
    gamesList.append(pd.read_pickle(filesGames[i]))
    gameDataList.append(pd.read_pickle(filesGameData[i]))
```

```python
# Merge lists

#name of columns
linksCol = "links"
gameNoCol = "game_no"
moveNoCol = "move_no"

# Set simple index
for i in range(len(gameDataList)-1):
    gameDataList[i].reset_index(inplace=True)
    #Set links in game data frames
    gameDataList[i][linksCol] = ""
    for j in range(len(gamesList[i])-1):
        gameDataList[i].loc[gameDataList[i][gameNoCol] == j, linksCol] = gamesList[i].loc[j][linksCol]

#gameDataList[0].loc[range(5), linksCol]
#gamesAll = pd.concat(gamesList).drop_duplicates(subset=linksCol).reset_index(drop=True)

game_no ändern auf Basis des Links

#This does not work yet, i guess
#gameDataAll = pd.concat(gameDataList).drop_duplicates(subset=[linksCol, moveNoCol]).reset_index(drop=True)#.set_index([gameNoCol, moveNoCol], inplace=True)
gameDataAll = pd.concat(gameDataList)
gameDataAll
```
